```python
# ==========================================
# IMPORT LIBRARIES
# ==========================================

import pandas as pd
import numpy as np

print("="*60)
print("LIBRARIES IMPORTED SUCCESSFULLY")
print("="*60)
```

```python
# ==========================================
# CREATE SAMPLE DATASET
# ==========================================

data = {
    "ID":[101,102,103,104,105,105,106,107,108,109],
    "Name":["Amit","Rahul","Priya","Sneha","Amit","Amit","Neha",None,"Karan","Riya"],
    "Age":[20,21,np.nan,22,20,20,23,21,np.nan,24],
    "City":["Ahmedabad","Surat","Rajkot","Ahmedabad","Ahmedabad","Ahmedabad","Vadodara","Rajkot","Surat",None],
    "Marks":[80,90,np.nan,75,80,80,95,88,92,90]
}

df = pd.DataFrame(data)

print("="*60)
print("ORIGINAL DATASET")
print("="*60)

display(df)
```

```python
# ==========================================
# DATASET INFORMATION
# ==========================================

print("="*60)
print("DATASET INFORMATION")
print("="*60)

df.info()

print("\nShape :", df.shape)

print("\nColumns")
print(df.columns.tolist())

print("\nData Types")
print(df.dtypes)
```

```python
# ==========================================
# MISSING VALUES
# ==========================================

print("="*60)
print("MISSING VALUES")
print("="*60)

print(df.isnull().sum())
```

```python
# ==========================================
# DUPLICATE ROWS
# ==========================================

print("="*60)
print("DUPLICATE ROWS")
print("="*60)

duplicates = df[df.duplicated()]

display(duplicates)

print("Total Duplicate Rows :", df.duplicated().sum())
```

```python
# ==========================================
# REPEATED VALUES
# ==========================================

print("="*60)
print("REPEATED NAMES")
print("="*60)

print(df["Name"].value_counts())

print()

print("="*60)
print("REPEATED CITIES")
print("="*60)

print(df["City"].value_counts())
```

```python
# ==========================================
# HANDLE MISSING VALUES
# ==========================================

df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

df["Name"] = df["Name"].fillna("Unknown")

df["City"] = df["City"].fillna("Unknown")

print("="*60)
print("MISSING VALUES FILLED")
print("="*60)

display(df)
```

```python
# ==========================================
# REMOVE DUPLICATE ROWS
# ==========================================

df = df.drop_duplicates()

print("="*60)
print("DUPLICATE ROWS REMOVED")
print("="*60)

print("New Shape :", df.shape)

display(df)
```

```python
# ==========================================
# REMOVE EXTRA SPACES
# ==========================================

df["Name"] = df["Name"].str.strip()

df["City"] = df["City"].str.strip()

print("="*60)
print("EXTRA SPACES REMOVED")
print("="*60)
```

```python
# ==========================================
# DATA TYPE CONVERSION
# ==========================================

df["Age"] = df["Age"].astype(float)

df["Marks"] = df["Marks"].astype(float)

print("="*60)
print("DATA TYPES AFTER CONVERSION")
print("="*60)

print(df.dtypes)
```

```python
# ==========================================
# DATA STATISTICS
# ==========================================

print("="*60)
print("DESCRIPTIVE STATISTICS")
print("="*60)

display(df.describe())
```

```python
# ==========================================
# FINAL CLEAN DATASET
# ==========================================

print("="*60)
print("FINAL CLEAN DATASET")
print("="*60)

display(df)
```

```python
# ==========================================
# VERIFY CLEAN DATA
# ==========================================

print("="*60)
print("FINAL VERIFICATION")
print("="*60)

print("\nMissing Values")

print(df.isnull().sum())

print()

print("Duplicate Rows")

print(df.duplicated().sum())

print()

print("Shape")

print(df.shape)

print()

print("Columns")

print(df.columns.tolist())

print()

print("Data Types")

print(df.dtypes)

print()

print("="*60)
print("DATA CLEANING COMPLETED SUCCESSFULLY")
print("="*60)
```
